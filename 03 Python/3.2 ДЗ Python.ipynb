0) Для генерации данных в random необходимо установить seed = "АДВИ_2023", длина каждого списка N = 1000;

```python
import pandas as pd
import random

random.seed("АДВИ_2023")

N = 1000
```

1) index колонка: создать список из N последовательных целых элементов от 1 до 1000 включительно, далее, с помощью специальной функции random перемешать его;

```python
# Создаем список N последовательных целых элементов от 1 до 1000
index = list(range(1, N+1))

# Перемешиваем список с помощью функции shuffle
random.shuffle(index)
```

2) groups колонка: создать список из N элементов, состоящий из значений из списка ["группа_1", "группа_2", "группа_3", "группа_4", "группа_5"], необходимо использовать функцию из random которая возвращает список элементов длины k, выбранных из последовательности population с перестановкой элементов;

```python
k = 1  # длина списка, который вернет функция sample
population = ["группа_1", "группа_2", "группа_3", "группа_4", "группа_5"]

# Создаем список groups с помощью функции sample
groups = random.choices(population, k=N)
# print(groups)
```

3) uniform колонка: создать список из N элементов, состоящий из целых значение из равномерного распределения [50, 100] ;

```python
# Создаем список uniform из N целых значений, выбранных из равномерного распределения [50, 100]
uniform = [random.randint(50, 100) for i in range(N)]

# print(uniform)
```

4) gauss_1, gauss_2, gauss_3 колонки: Создать 3 списка из N элементов каждый, заполненные значениями из нормальных распределений с параметрами:
- μ = 0, σ**2 = 1 -> третья колонка gauss_1;
- μ = 0, σ**2 = 144 -> четвертая колонка gauss_2;
- μ = 50, σ**2 = 81 -> пятая колонка gauss_3;

```python
# Создаем список gauss_1 из N элементов, выбранных из нормального распределения с параметрами μ = 0, σ^2 = 1
gauss_1 = [random.gauss(0, 1) for i in range(N)]

# Создаем список gauss_2 из N элементов, выбранных из нормального распределения с параметрами μ = 0, σ^2 = 144
gauss_2 = [random.gauss(0, 12) for i in range(N)]

# Создаем список gauss_3 из N элементов, выбранных из нормального распределения с параметрами μ = 50, σ^2 = 81
gauss_3 = [random.gauss(50, 9) for i in range(N)]

# print(gauss_1)
# print(gauss_2)
# print(gauss_3)
```

5) Создать pandas DataFrame, где (1) будет индексом, (2)-(4) колонки, названия колонок ["groups", "uniform", "gauss_1", "gauss_2", "gauss_3"] соответственно, не забывая проверить, что типы колонок должны соответствовать заявленным типам данных;

```python
# Создаем DataFrame с помощью метода DataFrame библиотеки pandas
df = pd.DataFrame(
    {
        "groups": groups,
        "uniform": uniform,
        "gauss_1": gauss_1,
        "gauss_2": gauss_2,
        "gauss_3": gauss_3,
    },
    index=index,
)

print(df.dtypes)

df.head()
```

6) Для колонок gauss_1 и gauss_2 заменить значения на None:

Для gauss_1, если индекс у значение делится на 121 без остатка;
Для gauss_2, если дробная часть больше 0.95;


```python
# Заменяем значения в колонке gauss_1 на None для индексов, делящихся на 121 без остатка
df.loc[df.index % 121 == 0, "gauss_1"] = None

# Заменяем значения в колонке gauss_2 на None для значений, у которых дробная часть больше 0.95
df.loc[df["gauss_2"].apply(lambda x: x % 1 > 0.95), "gauss_2"] = None

# Выводим первые 5 строк DataFrame для проверки
df.head()
```

7) Теперь, для пропущенных значений gauss_1 и gauss_2:

Для gauss_2 None заменить на среднее по колонке;
Удалить строчки из таблицы, где gauss_1 == None.

```python
# Заменяем пропущенные значения в колонке gauss_2 на среднее по колонке
df["gauss_2"].fillna(df["gauss_2"].mean(), inplace=True)

# Удаляем строки из таблицы, где gauss_1 == None
df.dropna(subset=["gauss_1"], inplace=True)

# Выводим первые 5 строк DataFrame для проверки
df.head()
df.info()
```
